# **Drop null records**

```python
df = df.dropna()
```

```python
index_drop_ProdYear_0 = df[(df['Prod. year'] == 0)].index
df = df.drop(index_drop_ProdYear_0, axis = 0)
```

```python
df["Levy"]=df["Levy"].replace('-',0)
df["Levy"]=df["Levy"].astype(float)
```

```python
Engine_volume = df['Engine volume'].apply(lambda x : x.split(' ')[0])
df.insert(3, "Engine_volume", Engine_volume)
df.drop(['Engine volume'], axis=1, inplace=True)
df["Engine_volume"]=df["Engine_volume"].astype(float)
```

```python
Mileage_f = df['Mileage'].apply(lambda x : x.split(' ')[0])
df.insert(3, "Mileage_f", Mileage_f)
df.drop(['Mileage'], axis=1, inplace=True)
df["Mileage_f"]=df["Mileage_f"].astype(float)
```

```python
# Extracting categorical and numerical columns

cat_cols = [col for col in df.columns if (df[col].dtype == 'object' or df[col].dtype == 'category')]
num_cols = [col for col in df.columns if (df[col].dtype != 'object' and df[col].dtype != 'category')]
```

```python
df.columns
```

```python
le = LabelEncoder()

for col in range(len(cat_cols)):
    new_col_name = str(cat_cols[col])+'_label_encoded'
    df[new_col_name] = le.fit_transform(df[cat_cols[col]])
    df.drop([cat_cols[col]], axis=1, inplace=True)

  
```
